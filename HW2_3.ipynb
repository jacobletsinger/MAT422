### 2.3.1 Joint probability distributions

When considering two discrete random variables the joint probability distribution can be used to find to probability of two events happening together. This is possible in python with seaborn, pandas, and numpy and is done here with two random variables of 100 random values.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

n=100
x = np.random.rand(n)
y = np.random.rand(n)

plt.hist2d(x,y)
plt.title('x vs. y')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
joint_prob = pd.DataFrame({'x': x, 'y': y})

sns.jointplot(x='x',y='y',data=joint_prob)
plt.title('Joint Probability')
plt.show()
```

### 2.3.2. Correlation and dependence
The correlation of two random variables is a measure of how strongly related they are. Below, $x$ and $y$ are strongly correlated while $x$ and $w$ are not. The correlation between random variables can be done in python with seaborn, pandas and numpy, and is done here with six random variables where $x$ and $y$ have been forced to be correlated by making $y$ a linearly related to $x$.  

```python
n=100
x = np.random.rand(n)
y = 2*x - np.random.rand(n)
w = np.random.rand(n)
p = np.random.rand(n)
q = np.random.rand(n)
r = np.random.rand(n)

plt.scatter(x,y)
plt.title('x vs. y')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

plt.scatter(x,w)
plt.title('x vs. w')
plt.xlabel('x')
plt.ylabel('w')
plt.show()
```

```python
df = pd.DataFrame({'x': x, 'y': y, 'w': w, 'p': p, 'q': q, 'r': r})

corr = df.corr()

sns.heatmap(corr)
plt.title("Correlation")
plt.show()
```

### 2.3.3. Random samples
Random samples are variables made of a certain sample size made up of that many random values in a certain interval. In python this is possible with numpy: np.random.rand(n) gives n numbers between 0 and 1. The central limit theorem state that the distribution of the mean of a independent random variables approaches a normal distribution as the sample size increases, which is illustrated with plots of different sample size.

```python
print(np.random.rand(5))

def clt(n):
  data = np.random.uniform(low=0, high=1, size=n * 100)

  data_samples = data.reshape((n, 100))

  sample_means = np.mean(data_samples, axis=1)

  plt.figure(figsize=(10, 6))

  plt.subplot(1, 2, 1)
  sns.histplot(data, color='blue')
  plt.title('Original Uniform Distribution')

  plt.subplot(1, 2, 2)
  sns.histplot(sample_means, color='green')
  plt.title('Distribution of Means (CLT)')

  plt.tight_layout()
  plt.show()

clt(10)
```

```python
clt(100)
```

```python
clt(1000)
```

```python
clt(10000)
```
